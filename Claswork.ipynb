```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

plt.rcParams['figure.figsize']=[12,10]
sns.set_style('darkgrid')

titanic_data = sns.load_dataset('titanic')
feature = ['survived','pclass','age','fare']
titanic_data1 = titanic_data[feature]
titanic_data.head()
```

```python
titanic_data1.head()
```

```python
is_null = titanic_data1.isnull().mean()
mean = titanic_data1.age.mean()
median = titanic_data1.age.median()
print(is_null)
print(f'Mean i s-- {mean} and Median is -- {median}')
```

```python
titanic_data1['Median_age'] = titanic_data1.age.fillna(median)
titanic_data1['Mean_age'] = titanic_data1.age.fillna(mean)



titanic_data1.head(20)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
titanic_data1['age'].plot(kind='kde',ax = ax)
titanic_data1['Median_age'].plot(kind='kde',ax = ax,color='red')
titanic_data1['Mean_age'].plot(kind='kde',ax = ax, color='green')
lines, labels =ax.get_legend_handles_labels()
ax.legend(lines,labels,loc = 'best')
```
